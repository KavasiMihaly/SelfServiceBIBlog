```python
##Required packages for demo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Put file path here
df = pd.read_csv(r'file')
```

```python
##Data transformation examples
df['Date'] = pd.to_datetime(df['Date'])
df = df.query('CustomerID != "Customer 999999"')
```

```python
print(df)
```

```python
print(df['Profit'])
```

```python
#Data visualization example
xp = df['Profit']
yp = df['Sales']

plt.scatter(xp,yp, c=xp, cmap = 'RdBu', s = 75, edgecolors = 'black')
plt.title("Sales vs Profit Analysis")
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
